```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from deepal6 import ActiveLearner, TabularDataModule, ALConfig
```

```python
# Synthetic data (replace with your real dataset)
X, y = make_classification(n_samples=800, n_features=15,
                            weights=[0.7, 0.3], random_state=42)
X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.2,
                                            stratify=y, random_state=42)
scaler = StandardScaler()
X_tr = scaler.fit_transform(X_tr)
X_te = scaler.transform(X_te)

# Wrap data
data = TabularDataModule(X_tr, y_tr, X_te, y_te)

# Configure (fast settings for a quick check)
cfg = ALConfig(
    strategy     = ['Random', 'Entropy', 'BALD'],
    initial_size = 50,
    batch_size   = 20,
    n_rounds     = 10,
    n_seeds      = 3,
    train_epochs = 30,
)

# Run
learner  = ActiveLearner(data, cfg)
results  = learner.run()

# View results
learner.summary_table(results, metric='auc')
learner.plot(results, metric='auc')
```

```python
from deepal6 import ActiveLearner, ImageDataModule, ALConfig
from torchvision import transforms

from torchvision import transforms
import medmnist
from medmnist import INFO
import numpy as np
from deepal6 import ImageDataModule

info   = INFO['bloodmnist']
DClass = getattr(medmnist, info['python_class'])

# Default transform — converts PIL → tensor and normalises for ResNet
_default_transform = transforms.Compose([
    transforms.Grayscale(num_output_channels=3),   # MedMNIST is grayscale
    transforms.Resize((64, 64)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225]),
])

class MedMNISTBinary(DClass):
    def __init__(self, split):
        super().__init__(split=split, download=True)
        self._labels    = (self.labels.squeeze() >= 4).astype(int)
        self._transform = _default_transform       # ← must convert PIL → tensor

    def get_labels(self):
        return self._labels

    def __getitem__(self, idx):
        img, _ = super().__getitem__(idx)          # PIL Image
        img    = self._transform(img)              # ← tensor now
        return img, int(self._labels[idx])

data = ImageDataModule(
    MedMNISTBinary('train'),
    MedMNISTBinary('test'),
    pos_label   = 1,
    num_workers = 0,
)

# Optional augmentations
my_aug = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),

    # IMPORTANT for pretrained ResNet
    transforms.Grayscale(num_output_channels=3),

    transforms.ToTensor(),

    transforms.Normalize(
        [0.485, 0.456, 0.406],
        [0.229, 0.224, 0.225]
    ),
])

config = ALConfig(
    strategy     = ['Random', 'CoreSet', 'BADGE'],
    initial_size = 50,
    batch_size   = 20,
    n_rounds     = 20,
    n_seeds      = 2,
    train_epochs = 10,
    lr           = 1e-4,
    dropout_rate = 0.4,
)

learner = ActiveLearner(data, config)

results = learner.run()

learner.plot(results, metric='auc', show_std=True)
```
